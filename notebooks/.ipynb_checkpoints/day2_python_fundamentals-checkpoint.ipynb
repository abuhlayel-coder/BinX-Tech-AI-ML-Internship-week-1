## Python Data Types

This section shows the basic Python data types used in programming and in data science.

```python
age = 21 #int
height =1.78 #float
name ="abdelrahman" #str
is_std = True #bool
marks = [75,80,85,90] #list
coordinates = (50,30) #tuple
student = {
    "name":"abdelrahman",
    "age":21,
    "year":"3rd"
} #dict
unique_numbers = {1,2,2,3,3,4} #set
print(type(age))
print(age)
print(type(height))
print(height)
print(type(name))
print(name)
print(type(is_std))
print(is_std)
print(type(marks))
print(marks)
print(type(coordinates))
print(coordinates)
print(type(student))
print(student)
print(type(unique_numbers))
print(unique_numbers)
```

## Control Flow

Python uses if statements and loops to control program execution.

```python
number = 5

if number % 2 == 0: 
   print("this number is even")
else:
   print("this number is odd")
print()

for i in range(7): 
    print(i)

print()
count = -1
while count <=5:
    print(count)
    count += 1

for value in [1, 2, 3, 4]:
    if value % 2 == 0:
        print(f"{value} is even")
    else:
        print(f"{value} is odd")
```

## Functions

Functions allow reusable code.

```python
def my_function(country = "Palestain"):
  print("I am from", country)

my_function("Jordan")
my_function("Spain")
my_function()
my_function("Egypt")
```

## List Comprehensions

List comprehensions provide a shorter and more Pythonic way to create lists.

```python
countrys = ["Palestain", "Jordan", "Spain", "Egypt", "Turkey"]
newlist = []

for x in countrys:
  if "a" in x:
    newlist.append(x)

print(newlist)
```

## Object-Oriented Programming

Classes combine data and methods into reusable objects.

```python
class DataPoint:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def distance_from_origin(self):
        return (self.x**2 + self.y**2) ** 0.5
p = DataPoint(3, 4)
print(p.distance_from_origin())   # 5.0

class MyClass:
  x = 5

p1 = MyClass()
print(p1.x)
del p1


p2 = MyClass()
p3 = MyClass()


print(p2.x)
print(p3.x)

class Test:
  pass
```

## Step 1 - Function

Write a function that takes a list of numbers and returns the mean, min, and max as a dictionary:
This function receives a list of numbers and calculates the mean, minimum, and maximum values. 
Why its useful? 
Using functions improves code organization and reusability.

```python
def analyze_Function(numbers):
    result = {
        "mean": sum(numbers) / len(numbers),
        "min": min(numbers),
        "max": max(numbers)
    }
    return result

values = [20,5,17,18,10]

print(analyze_Function(values))
```

## Step 2 - List Comprehension

Rewrite a given for-loop that filters even numbers as a single list comprehension:
This example filters even numbers using a list comprehension
Why its useful?
More concise and Pythonic alternative to a standard for-loop.

```python
numbers = list(range(20))

even_numbers = [num for num in numbers if num % 2 == 0]

print(even_numbers)
```

## Step 3 - Object-Oriented Programming

Define a small class representing a data record with at least two attributes and one method:
This class represents a student record with two attributes and one method.
Why its useful?
It demonstrates the basics of object-oriented programming in Python.

```python
class Student:

    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def display_info(self):
        return f"Student: {self.name}, Grade: {self.grade}"

student1 = Student("abdelrahman", 95)

print(student1.display_info())
```
