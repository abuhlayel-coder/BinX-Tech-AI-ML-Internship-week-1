## Why visualization matters

A table of numbers hides patterns that a chart reveals instantly — trends, clusters, outliers, relationships between variables. It's both an exploration tool (for you, while analyzing) and a communication tool (for explaining findings to others). It's the foundation of the Exploratory Data Analysis (EDA) work coming in Week 2.

## Matplotlib Basics

The standard workflow: plot -> label axes -> title -> show. An unlabeled chart isn't a finding, it's a puzzle.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("A simple line plot")
plt.show()
```

## The Four Core Plot Types

- **Line** `plt.plot()` - a trend over a continuous axis
- **Scatter** `plt.scatter()` - relationship between two numeric variables
- **Bar** `plt.bar()` - comparison across categories
- **Histogram** `plt.hist()` - distribution of a single variable

```python
import numpy as np

categories = ["A", "B", "C", "D"]
values = [23, 45, 12, 36]

plt.bar(categories, values)
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Bar plot example")
plt.show()
```

```python
data = np.random.normal(loc=50, scale=10, size=500)

plt.hist(data, bins=20)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram example")
plt.show()
```

## Subplots

Subplots place multiple charts in one figure for side-by-side comparison, using `plt.subplots()` to create a grid of axes.

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].hist(data, bins=20)
axes[0].set_title("Distribution")

axes[1].scatter(np.arange(len(values)), values)
axes[1].set_title("Values by index")

plt.tight_layout()
plt.show()
```
