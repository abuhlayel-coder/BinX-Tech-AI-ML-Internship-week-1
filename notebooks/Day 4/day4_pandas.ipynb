## Selecting and Filtering

Select columns by name, and rows by condition (boolean filtering). `.loc` selects by label, `.iloc` by integer position.

```python
# single column (Series) and multiple columns (DataFrame)
ages = df["age"]
subset = df[["age", "fare", "class"]]
subset.head()
```

## Cleaning Data

Real data is messy: missing values, duplicates, and wrong types are the norm. Cleaning is often the largest part of any ML project, and doing it correctly is what everything downstream depends on.

```python
df.isnull().sum()
```

```python
df_clean = df.copy()

# fill missing numeric age with the median (robust to outliers)
df_clean["age"] = df_clean["age"].fillna(df_clean["age"].median())

# fill missing categorical embark_town/embarked with the most common value
df_clean["embark_town"] = df_clean["embark_town"].fillna(df_clean["embark_town"].mode()[0])
df_clean["embarked"] = df_clean["embarked"].fillna(df_clean["embarked"].mode()[0])

# drop deck: too many missing values to meaningfully impute
df_clean = df_clean.drop(columns=["deck"])

df_clean = df_clean.drop_duplicates()

print(df_clean.isnull().sum())
print("shape after cleaning:", df_clean.shape)
```

## Grouping and Aggregation

`groupby` splits the data into groups by a column's values, applies an aggregation to each group, and combines the results — the single most useful Pandas operation for finding patterns in data.

```python
df_clean.groupby("class")["survived"].mean()
```

```python
df_clean.groupby("class").agg({"fare": "mean", "age": "mean"})
```

## Step 1 - Load and Inspect

Load the Titanic dataset into a DataFrame and report its shape, columns, and dtypes.

```python
titanic = pd.read_csv("titanic.csv")

print("Shape:", titanic.shape)
print("\nColumns:", list(titanic.columns))
print("\nDtypes:\n", titanic.dtypes)
```

## Step 2 - Handle Missing Values

Count missing values per column, then fill or drop them with a documented justification.

```python
print(titanic.isnull().sum())
```

```python
# age: numeric, ~20% missing -> fill with median (robust to skew from outliers)
titanic["age"] = titanic["age"].fillna(titanic["age"].median())

# embarked / embark_town: only 2 missing -> fill with the most frequent port
titanic["embarked"] = titanic["embarked"].fillna(titanic["embarked"].mode()[0])
titanic["embark_town"] = titanic["embark_town"].fillna(titanic["embark_town"].mode()[0])

# deck: over 75% missing -> not enough signal to impute reliably, so it's dropped
titanic = titanic.drop(columns=["deck"])

titanic = titanic.drop_duplicates()

print("Missing values after cleaning:")
print(titanic.isnull().sum())
print("\nFinal shape:", titanic.shape)
```

**Justification:** `age` was filled with the median rather than the mean since fare/age distributions on the Titanic are right-skewed, and the median is less sensitive to outliers. `embarked`/`embark_town` had only 2 missing rows, so filling with the most common port introduces negligible bias. `deck` was dropped entirely because over three-quarters of its values were missing — imputing it would mostly be guessing.

## Step 3 - Filter to a Meaningful Subset

Filter the data to a meaningful subset and describe what's found.

```python
# Subset: female passengers in 1st or 2nd class
subset = titanic[(titanic["sex"] == "female") & (titanic["pclass"] <= 2)]

print("Number of passengers in subset:", len(subset))
print("Survival rate in subset:", round(subset["survived"].mean(), 3))
print("Overall survival rate:", round(titanic["survived"].mean(), 3))
```

**Finding:** Female passengers in 1st or 2nd class survived at a noticeably higher rate than the overall passenger average, consistent with the "women and children first" evacuation priority combined with better access to lifeboats from higher-class cabins.

## Step 4 - Group and Aggregate

Use groupby to compute an aggregate statistic per category and interpret the result.

```python
survival_by_class_sex = titanic.groupby(["class", "sex"])["survived"].mean()
survival_by_class_sex
```

**Interpretation:** Survival rate drops sharply moving from 1st to 3rd class within both sexes, and women outsurvived men in every class. The combination of class and sex is clearly one of the strongest predictors of survival in this dataset — exactly the kind of pattern `groupby` is built to surface.
